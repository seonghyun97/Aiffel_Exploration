## 03. 실험함수 만들기 🧪


---



train data를 가지고 훈련시키고 test해보며 model별 성능을 수치와 그래프를 활용하여 비교하여본다.

```python
'''
y_test, y_pred data의 rmse를 구해주는 function
'''
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

```python
# random_state는 모델초기화나 데이터셋 구성에 사용되는 랜덤 시드값입니다. 
#random_state=None    # 이게 초기값입니다. 아무것도 지정하지 않고 None을 넘겨주면 모델 내부에서 임의로 선택합니다.  
random_state=2020        # 하지만 우리는 이렇게 고정값을 세팅해 두겠습니다. 

# model 배정
gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)
```

```python
'''
각 model별 rmse 점수와 graph를 그려 직관적으로 확인한다.
'''

def get_scores(models, train, y):
    #score를 나타내줄 dataframe
    df = {}
    
    # graph index로 쓰임
    f = [i for i in range(100)]
    for model in models:
        # method를 사용하여 model name을 가져와서 사용할 수 있다.
        model_name = model.__class__.__name__
        
        # dataset 분리 및 훈련
        X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=random_state, test_size=0.2)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # model별 RMSE 배정
        y_ps = pd.Series(y_pred)
        df[model_name] = rmse(y_test, y_pred)
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
        
        # model 별 graph 그리기 data가 너무 많아 앞에 100개만 확인해본다.
        fig = plt.figure( figsize = (12, 4) ) 
        chart = fig.add_subplot(1,1,1) 
        chart.scatter(f, y_test[:100], color='blue', label='real value') 
        chart.scatter(f, y_ps[:100], color='red', label='predict value') 
        chart.set_title(model_name) 
        plt.xlabel('index') 
        plt.ylabel('real vs predict') 
        plt.legend(loc = 'best')
            
    return score_df


```

```python
''' models list에 있는 model들을 train data와 y data를 넣어서 성능을 비교하여본다.'''
models = [gboost, xgboost, lightgbm, rdforest]
get_scores(models, train, y)
```

- GradientBoostingRegressor	126311.883915
- RandomForestRegressor	123824.167476
- LGBMRegressor	112653.387024

RMSE 값과 Graph를 보았을 때에 LGBM, RandomForest, Gradient, XGB 순서로 성능이 좋을 것으로 예상이 된다.

## 04. 하이퍼 파라미터 튜닝


---



GridSearch를 활용하여 Hyper parameter를 최적으로 tuning한다. 다양한 범위를 정해주어 찾을 수 있도록 해준다.

```python
'''
GridSearchCV 활용

param_grid : 탐색할 파라미터의 종류 (딕셔너리로 입력)
scoring : 모델의 성능을 평가할 지표
cv : cross validation을 수행하기 위해 train 데이터셋을 나누는 조각의 개수
verbose : 그리드 탐색을 진행하면서 진행 과정을 출력해서 보여줄 메세지의 양 (숫자가 클수록 더 많은 메세지를 출력합니다.)
n_jobs : 그리드 탐색을 진행하면서 사용할 CPU의 개수

'''

"""
다음과 같은 과정을 진행할 수 있는 `my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5)` 함수를 구현해 보세요.

1. GridSearchCV 모델로 `model`을 초기화합니다.
2. 모델을 fitting 합니다.
3. params, score에 각 조합에 대한 결과를 저장합니다. 
4. 데이터 프레임을 생성하고, RMSLE 값을 추가한 후 점수가 높은 순서로 정렬한 `results`를 반환합니다.
"""

# 코드 입력
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):
    # GridSearchCV 모델로 초기화
    grid_model = GridSearchCV(model, param_grid=param_grid, scoring='neg_mean_squared_error', \
                              cv=5, verbose=verbose, n_jobs=n_jobs)
    
    # 모델 fitting
    grid_model.fit(train, y)

    # 결과값 저장
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    
    # 데이터 프레임 생성
    results = pd.DataFrame(params)
    results['score'] = score
    
    # RMSLE 값 계산 후 정렬
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results = results.sort_values('RMSLE')

    return results
```

### 04-1. LGBMR

```python
param_grid = {
    'objective':['regression'],
    'learning_rate' : [0.03],
    'n_estimators':[100,500,1000],
    'max_depth': [1, 10, 20, 30, 40, 50],
}

model = LGBMRegressor(random_state=random_state)
my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5)
```

**위의 결과들로 봤을 때에도 XGBR이 가장 높은 성능을 보여줄 것으로 예상된다.**

```python
"""
아래의 과정을 수행하는 `save_submission(model, train, y, test, model_name, rmsle)` 함수를 구현해 주세요.
1. 모델을 `train`, `y`로 학습시킵니다.
2. `test`에 대해 예측합니다.
3. 예측값을 `np.expm1`으로 변환하고, `submission_model_name_RMSLE_newway.csv` 형태의 `csv` 파일을 저장합니다.
"""

# 코드 작성
def save_submission(model, train, y, test, model_name, rmsle=None):
    model.fit(train, y)
    prediction = model.predict(test)
    prediction = np.expm1(prediction)
    data_dir = '/content/data'
    submission_path = join(data_dir, 'sample_submission.csv')
    submission = pd.read_csv(submission_path)
    submission['price'] = prediction
    submission_csv_path = '{}/submission_{}_RMSLE_{}.csv'.format(data_dir, model_name, rmsle)
    submission.to_csv(submission_csv_path, index=False)
    print('{} saved!'.format(submission_csv_path))
```

```python
# LGBM
model = LGBMRegressor(max_depth=10, n_estimators=100, learning_rate=0.03, objective='regression', random_state=random_state)
save_submission(model, train, y, test, 'lgbm', rmsle='newway')

# XGB
model = XGBRegressor(max_depth=10, n_estimators=500, learning_rate=0.05, subsample=0.9, random_state=random_state)
save_submission(model, train, y, test, 'xgb', rmsle='newway')

# RandomForest
model = RandomForestRegressor(max_depth=50, n_estimators=1000, random_state=random_state)
save_submission(model, train, y, test, 'rf', rmsle='newway')
```

- One-Hot encoding
>One-hot encoding 시도했을 때 feature의 개수가 너무 많아져 모델의 학습시간이 길어졌다. model을 돌리고 동네 한바퀴 산책을 하고와야 결과값이 나와 있었다. 이 feature 갯수를 어떻게 줄일 수 있었을까? 고민을 하다가 범주형 data를 category화 시켜서 하는 방법이 있다고 한다. 어떤 기준을 가지고 category를 만들어야할지 몰라서 시도는 하지 못했지만 One-Hot encoding의 단점인 feature 갯수가 너무 많아질 수 있다라는 것을 몸으로 느낄 수 있는 시간이었고 다음에 기회가 있으면 category화 시키는 것도 시도해봐야겠다.
