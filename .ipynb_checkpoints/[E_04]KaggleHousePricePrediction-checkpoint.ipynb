## 03. Ïã§ÌóòÌï®Ïàò ÎßåÎì§Í∏∞ üß™


---



train dataÎ•º Í∞ÄÏßÄÍ≥† ÌõàÎ†®ÏãúÌÇ§Í≥† testÌï¥Î≥¥Î©∞ modelÎ≥Ñ ÏÑ±Îä•ÏùÑ ÏàòÏπòÏôÄ Í∑∏ÎûòÌîÑÎ•º ÌôúÏö©ÌïòÏó¨ ÎπÑÍµêÌïòÏó¨Î≥∏Îã§.

```python
'''
y_test, y_pred dataÏùò rmseÎ•º Íµ¨Ìï¥Ï£ºÎäî function
'''
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

```python
# random_stateÎäî Î™®Îç∏Ï¥àÍ∏∞ÌôîÎÇò Îç∞Ïù¥ÌÑ∞ÏÖã Íµ¨ÏÑ±Ïóê ÏÇ¨Ïö©ÎêòÎäî ÎûúÎç§ ÏãúÎìúÍ∞íÏûÖÎãàÎã§. 
#random_state=None    # Ïù¥Í≤å Ï¥àÍ∏∞Í∞íÏûÖÎãàÎã§. ÏïÑÎ¨¥Í≤ÉÎèÑ ÏßÄÏ†ïÌïòÏßÄ ÏïäÍ≥† NoneÏùÑ ÎÑòÍ≤®Ï£ºÎ©¥ Î™®Îç∏ ÎÇ¥Î∂ÄÏóêÏÑú ÏûÑÏùòÎ°ú ÏÑ†ÌÉùÌï©ÎãàÎã§.  
random_state=2020        # ÌïòÏßÄÎßå Ïö∞Î¶¨Îäî Ïù¥Î†áÍ≤å Í≥†Ï†ïÍ∞íÏùÑ ÏÑ∏ÌåÖÌï¥ ÎëêÍ≤†ÏäµÎãàÎã§. 

# model Î∞∞Ï†ï
gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)
```

```python
'''
Í∞Å modelÎ≥Ñ rmse Ï†êÏàòÏôÄ graphÎ•º Í∑∏Î†§ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú ÌôïÏù∏ÌïúÎã§.
'''

def get_scores(models, train, y):
    #scoreÎ•º ÎÇòÌÉÄÎÇ¥Ï§Ñ dataframe
    df = {}
    
    # graph indexÎ°ú Ïì∞ÏûÑ
    f = [i for i in range(100)]
    for model in models:
        # methodÎ•º ÏÇ¨Ïö©ÌïòÏó¨ model nameÏùÑ Í∞ÄÏ†∏ÏôÄÏÑú ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎã§.
        model_name = model.__class__.__name__
        
        # dataset Î∂ÑÎ¶¨ Î∞è ÌõàÎ†®
        X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=random_state, test_size=0.2)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # modelÎ≥Ñ RMSE Î∞∞Ï†ï
        y_ps = pd.Series(y_pred)
        df[model_name] = rmse(y_test, y_pred)
        score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
        
        # model Î≥Ñ graph Í∑∏Î¶¨Í∏∞ dataÍ∞Ä ÎÑàÎ¨¥ ÎßéÏïÑ ÏïûÏóê 100Í∞úÎßå ÌôïÏù∏Ìï¥Î≥∏Îã§.
        fig = plt.figure( figsize = (12, 4) ) 
        chart = fig.add_subplot(1,1,1) 
        chart.scatter(f, y_test[:100], color='blue', label='real value') 
        chart.scatter(f, y_ps[:100], color='red', label='predict value') 
        chart.set_title(model_name) 
        plt.xlabel('index') 
        plt.ylabel('real vs predict') 
        plt.legend(loc = 'best')
            
    return score_df


```

```python
''' models listÏóê ÏûàÎäî modelÎì§ÏùÑ train dataÏôÄ y dataÎ•º ÎÑ£Ïñ¥ÏÑú ÏÑ±Îä•ÏùÑ ÎπÑÍµêÌïòÏó¨Î≥∏Îã§.'''
models = [gboost, xgboost, lightgbm, rdforest]
get_scores(models, train, y)
```

## 04. ÌïòÏù¥Ìçº ÌååÎùºÎØ∏ÌÑ∞ ÌäúÎãù


---



GridSearchÎ•º ÌôúÏö©ÌïòÏó¨ Hyper parameterÎ•º ÏµúÏ†ÅÏúºÎ°ú tuningÌïúÎã§. Îã§ÏñëÌïú Î≤îÏúÑÎ•º Ï†ïÌï¥Ï£ºÏñ¥ Ï∞æÏùÑ Ïàò ÏûàÎèÑÎ°ù Ìï¥Ï§ÄÎã§.

```python
'''
GridSearchCV ÌôúÏö©

param_grid : ÌÉêÏÉâÌï† ÌååÎùºÎØ∏ÌÑ∞Ïùò Ï¢ÖÎ•ò (ÎîïÏÖîÎÑàÎ¶¨Î°ú ÏûÖÎ†•)
scoring : Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÌèâÍ∞ÄÌï† ÏßÄÌëú
cv : cross validationÏùÑ ÏàòÌñâÌïòÍ∏∞ ÏúÑÌï¥ train Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ ÎÇòÎàÑÎäî Ï°∞Í∞ÅÏùò Í∞úÏàò
verbose : Í∑∏Î¶¨Îìú ÌÉêÏÉâÏùÑ ÏßÑÌñâÌïòÎ©¥ÏÑú ÏßÑÌñâ Í≥ºÏ†ïÏùÑ Ï∂úÎ†•Ìï¥ÏÑú Î≥¥Ïó¨Ï§Ñ Î©îÏÑ∏ÏßÄÏùò Ïñë (Ïà´ÏûêÍ∞Ä ÌÅ¥ÏàòÎ°ù Îçî ÎßéÏùÄ Î©îÏÑ∏ÏßÄÎ•º Ï∂úÎ†•Ìï©ÎãàÎã§.)
n_jobs : Í∑∏Î¶¨Îìú ÌÉêÏÉâÏùÑ ÏßÑÌñâÌïòÎ©¥ÏÑú ÏÇ¨Ïö©Ìï† CPUÏùò Í∞úÏàò

'''

"""
Îã§ÏùåÍ≥º Í∞ôÏùÄ Í≥ºÏ†ïÏùÑ ÏßÑÌñâÌï† Ïàò ÏûàÎäî `my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5)` Ìï®ÏàòÎ•º Íµ¨ÌòÑÌï¥ Î≥¥ÏÑ∏Ïöî.

1. GridSearchCV Î™®Îç∏Î°ú `model`ÏùÑ Ï¥àÍ∏∞ÌôîÌï©ÎãàÎã§.
2. Î™®Îç∏ÏùÑ fitting Ìï©ÎãàÎã§.
3. params, scoreÏóê Í∞Å Ï°∞Ìï©Ïóê ÎåÄÌïú Í≤∞Í≥ºÎ•º Ï†ÄÏû•Ìï©ÎãàÎã§. 
4. Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ ÏÉùÏÑ±ÌïòÍ≥†, RMSLE Í∞íÏùÑ Ï∂îÍ∞ÄÌïú ÌõÑ Ï†êÏàòÍ∞Ä ÎÜíÏùÄ ÏàúÏÑúÎ°ú Ï†ïÎ†¨Ìïú `results`Î•º Î∞òÌôòÌï©ÎãàÎã§.
"""

# ÏΩîÎìú ÏûÖÎ†•
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):
    # GridSearchCV Î™®Îç∏Î°ú Ï¥àÍ∏∞Ìôî
    grid_model = GridSearchCV(model, param_grid=param_grid, scoring='neg_mean_squared_error', \
                              cv=5, verbose=verbose, n_jobs=n_jobs)
    
    # Î™®Îç∏ fitting
    grid_model.fit(train, y)

    # Í≤∞Í≥ºÍ∞í Ï†ÄÏû•
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    
    # Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
    results = pd.DataFrame(params)
    results['score'] = score
    
    # RMSLE Í∞í Í≥ÑÏÇ∞ ÌõÑ Ï†ïÎ†¨
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results = results.sort_values('RMSLE')

    return results
```

### 04-1. LGBMR

```python
param_grid = {
    'objective':['regression'],
    'learning_rate' : [0.03],
    'n_estimators':[100,500,1000],
    'max_depth': [1, 10, 20, 30, 40, 50],
}

model = LGBMRegressor(random_state=random_state)
my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5)
```

**ÏúÑÏùò Í≤∞Í≥ºÎì§Î°ú Î¥§ÏùÑ ÎïåÏóêÎèÑ XGBRÏù¥ Í∞ÄÏû• ÎÜíÏùÄ ÏÑ±Îä•ÏùÑ Î≥¥Ïó¨Ï§Ñ Í≤ÉÏúºÎ°ú ÏòàÏÉÅÎêúÎã§.**

```python
"""
ÏïÑÎûòÏùò Í≥ºÏ†ïÏùÑ ÏàòÌñâÌïòÎäî `save_submission(model, train, y, test, model_name, rmsle)` Ìï®ÏàòÎ•º Íµ¨ÌòÑÌï¥ Ï£ºÏÑ∏Ïöî.
1. Î™®Îç∏ÏùÑ `train`, `y`Î°ú ÌïôÏäµÏãúÌÇµÎãàÎã§.
2. `test`Ïóê ÎåÄÌï¥ ÏòàÏ∏°Ìï©ÎãàÎã§.
3. ÏòàÏ∏°Í∞íÏùÑ `np.expm1`ÏúºÎ°ú Î≥ÄÌôòÌïòÍ≥†, `submission_model_name_RMSLE_newway.csv` ÌòïÌÉúÏùò `csv` ÌååÏùºÏùÑ Ï†ÄÏû•Ìï©ÎãàÎã§.
"""

# ÏΩîÎìú ÏûëÏÑ±
def save_submission(model, train, y, test, model_name, rmsle=None):
    model.fit(train, y)
    prediction = model.predict(test)
    prediction = np.expm1(prediction)
    data_dir = '/content/data'
    submission_path = join(data_dir, 'sample_submission.csv')
    submission = pd.read_csv(submission_path)
    submission['price'] = prediction
    submission_csv_path = '{}/submission_{}_RMSLE_{}.csv'.format(data_dir, model_name, rmsle)
    submission.to_csv(submission_csv_path, index=False)
    print('{} saved!'.format(submission_csv_path))
```

```python
# LGBM
model = LGBMRegressor(max_depth=10, n_estimators=100, learning_rate=0.03, objective='regression', random_state=random_state)
save_submission(model, train, y, test, 'lgbm', rmsle='newway')

# XGB
model = XGBRegressor(max_depth=10, n_estimators=500, learning_rate=0.05, subsample=0.9, random_state=random_state)
save_submission(model, train, y, test, 'xgb', rmsle='newway')

# RandomForest
model = RandomForestRegressor(max_depth=50, n_estimators=1000, random_state=random_state)
save_submission(model, train, y, test, 'rf', rmsle='newway')
```
